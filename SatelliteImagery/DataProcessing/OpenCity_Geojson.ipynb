```python
df_acc = gpd.read_file(acc_geojson)
df_dar = gpd.read_file(dar_geojson)
df_mon = gpd.read_file(mon_geojson)
df_kam = gpd.read_file(kam_geojson)
df_ptn = gpd.read_file(ptn_geojson)
df_znz = gpd.read_file(znz_geojson)
```
